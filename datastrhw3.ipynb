```python
#question no 1
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def add_node(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            new_node.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.head

    def find_safest_nest(self):
        temp = self.head
        safest_nest = temp.data
        while temp.next != self.head:
            if temp.data < safest_nest:
                safest_nest = temp.data
            temp = temp.next
        if temp.data < safest_nest:
            safest_nest = temp.data
        return safest_nest

nests = CircularLinkedList()
for i in range(1, 11):
    nests.add_node(i)

safest_nest = nests.find_safest_nest()

print("The hare's safest nest is:", safest_nest)
```

```python
#question no.2
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def insert(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.tail.next = self.head

    def traverse(self):
        current_node = self.head
        if not current_node:
            print("Empty list.")
            return
        print(current_node.data, end=' ')
        current_node = current_node.next
        while current_node != self.head:
            print(current_node.data, end=' ')
            current_node = current_node.next
        print()

def common_nodes(l, m):
    common_vals = set()
    if not l.head or not m.head:
        return None
    l_current = l.head
    m_current = m.head
    while True:
        if l_current.data == m_current.data:
            common_vals.add(l_current.data)
        l_current = l_current.next
        m_current = m_current.next
        if l_current == l.head or m_current == m.head:
            break
    if len(common_vals) == 0:
        return None
    result = CircularLinkedList()
    for val in sorted(common_vals):
        result.insert(val)
    return result

# Example usage:
l = CircularLinkedList()
m = CircularLinkedList()
l.insert('a')
l.insert('b')
l.insert('d')
l.insert('e')
l.insert('f')
m.insert('a')
m.insert('c')
m.insert('d')
m.insert('e')
m.insert('g')
print("List l:")
l.traverse()
print("List m:")
m.traverse()
common_list = common_nodes(l, m)
if common_list:
    print("Common nodes in l and m:")
    common_list.traverse()
else:
    print("No common nodes found.")
```

```python
#question no 4
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
        else:
            current_node = self.head
            while current_node.next != self.head:
                current_node = current_node.next
            current_node.next = new_node
            new_node.next = self.head

    def print_list(self):
        if not self.head:
            print("List is empty")
        else:
            current_node = self.head
            print(current_node.data, end=" ")
            while current_node.next != self.head:
                current_node = current_node.next
                print(current_node.data, end=" ")
            print()

def split_CLL(org, n):
    current_node = org.head
    for i in range(n - 1):
        current_node = current_node.next
    new_head = current_node.next
    current_node.next = org.head
    new_list = CircularLinkedList()
    new_list.head = new_head
    new_node = new_head
    while new_node.next != org.head:
        new_node = new_node.next
    new_node.next = new_head
    return org, new_list

# Example usage
circular_list = CircularLinkedList()
circular_list.insert(1)
circular_list.insert(2)
circular_list.insert(3)
circular_list.insert(4)
circular_list.insert(5)
circular_list.insert(6)

print("Original List:")
circular_list.print_list()

list1, list2 = split_CLL(circular_list, 3)

print("List 1:")
list1.print_list()

print("List 2:")
list2.print_list()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None
 
def checkPointer(head):
    # Check if head is None
    if head is None:
        return False
 
    temp = head
 
    # Check if head's prev pointer is None
    if head.prev is not None:
        return False
     
    # Check for remaining nodes
    while temp is not None:
        # If node's prev pointer is None, return False
        if temp.prev is None:
            return False
         
        # If node's next pointer is None, return False
        if temp.next is None:
            return False
 
        temp = temp.next
     
    # If all pointers are correct, return True
    return True
 
def correctPointer(head):

    if head == None:
        return

    temp = head
    incorrect_pointers = []

    # if head.next's previous is not
    # pointing to head itself, change it.
    if (head.next != None and
        head.next.prev != head):

        head.next.prev = head
        incorrect_pointers.append((head, head.next))

    # If head's previous pointer is
    # incorrect, change it.
    if head.prev != None:

        head.prev = None
        incorrect_pointers.append((head, head.prev))

    # Else check for remaining nodes.
    temp = temp.next
    while temp != None:

        # If node.next's previous pointer
        # is incorrect, change it.
        if (temp.next != None and
            temp.next.prev != temp):

            temp.next.prev = temp
            incorrect_pointers.append((temp, temp.next))

        # Else If node.prev's next pointer
        # is incorrect, change it.
        elif (temp.prev != None and
              temp.prev.next != temp):

            temp.prev.next = temp
            incorrect_pointers.append((temp, temp.prev))

        # Else iterate on remaining.
        temp = temp.next

    if incorrect_pointers:
        print("The following pointers are incorrect:")
        for pair in incorrect_pointers:
            print(f"{pair[0].data} <-> {pair[1].data}")
    else:
        print("All pointers are correct.")

     
# Function to print the DLL
def printList(head):
    temp = head
 
    while temp is not None:
        print(temp.data, "(", end="")
 
        # If prev pointer is null, print -1.
        if temp.prev is None:
            print(-1, end=") ")
        else:
            print(temp.prev.data, end=") ")
         
        temp = temp.next
     
    print()
 
# Driver Code
if __name__ == "__main__":
 
    # Creating a DLL
    head = Node(1)
    head.next = Node(2)
    head.next.prev = head
    head.next.next = Node(3)
    head.next.next.prev = head
    head.next.next.next = Node(4)
    head.next.next.next.prev = head.next.next
 
    print("Incorrect Linked List:", end=" ")
    printList(head)
 
    # Check if pointers are correct
    if checkPointer(head):
        print("All pointers are correct")
    else:
        print("Incorrect pointers")
 
    # Correct pointers if necessary
    correctPointer(head)
 
    print("Corrected Linked List:", end=" ")
    printList(head)
```
